```python
import pandas as pd
import os

cr_src=pd.concat(
    [
        pd.read_csv(os.path.join('./cropped_UTK.csv')),
        pd.read_csv(os.path.join('./uncropped_UTK.csv'))
    ]
)

df_cropped_p_mtcnn = pd.read_csv(os.path.join('./crop_df_p_mtcnn.csv'))
df_cropped_p_mtcnn=pd.merge(left=cr_src,how='left',right=df_cropped_p_mtcnn)
df_cropped_p_mtcnn['detection_enabled']=True 
df_cropped_p_mtcnn['detection_model'] = 'mtcnn'
df_cropped_p_mtcnn['image_type'] = 'cropped'

df_uncropped_p_mtcnn = pd.read_csv(os.path.join('./uncropped_df_p_mtcnn.csv'))
df_uncropped_p_mtcnn=pd.merge(left=cr_src,how='left',right=df_uncropped_p_mtcnn)
df_uncropped_p_mtcnn['detection_enabled']=True 
df_uncropped_p_mtcnn['detection_model'] = 'mtcnn'
df_uncropped_p_mtcnn['image_type'] = 'uncropped'

df_cropped_p_opencv = pd.read_csv(os.path.join('./crop_df_p_opencv.csv'))
df_cropped_p_opencv=pd.merge(left=cr_src,how='left',right=df_cropped_p_opencv)
df_cropped_p_opencv['detection_enabled']=True 
df_cropped_p_opencv['detection_model'] = 'opencv'
df_cropped_p_opencv['image_type'] = 'cropped'

df_uncropped_p_opencv = pd.read_csv(os.path.join('./uncropped_df_p_opencv.csv'))
df_uncropped_p_opencv=pd.merge(left=cr_src,how='left',right=df_uncropped_p_opencv)
df_uncropped_p_opencv['detection_enabled']=True 
df_uncropped_p_opencv['detection_model'] = 'opencv'
df_uncropped_p_opencv['image_type'] = 'uncropped'

df_cropped_np = pd.read_csv(os.path.join('./crop_df_np.csv'))
df_cropped_np=pd.merge(left=cr_src,how='left',right=df_cropped_np)
df_cropped_np['detection_enabled']=False
df_cropped_np['detection_model'] = None
df_cropped_np['image_type'] = 'cropped'

df_uncropped_np = pd.read_csv(os.path.join('./uncropped_df_np.csv'))
df_uncropped_np=pd.merge(left=cr_src,how='left',right=df_uncropped_np)
df_uncropped_np['detection_enabled']=False
df_uncropped_np['detection_model'] = None
df_uncropped_np['image_type'] = 'uncropped'

ff_cropped_p = pd.read_csv(os.path.join('./cropped_ff_p.csv'))
ff_cropped_p=pd.merge(left=cr_src,how='left',right=ff_cropped_p)
ff_cropped_p['detection_enabled']=True
ff_cropped_p['detection_model'] = 'dlib'
ff_cropped_p['image_type'] = 'cropped'
ff_cropped_p=ff_cropped_p[['file','pred_race', 'pred_gender', 'pred_age_grp',
       'pred_age_lower', 'pred_age_upper', 'pred_model', 'detection_enabled',
       'detection_model', 'image_type']]

ff_uncropped_p = pd.read_csv(os.path.join('./uncropped_ff_p.csv'))
ff_uncropped_p=pd.merge(left=cr_src,how='left',right=ff_uncropped_p)
ff_uncropped_p['detection_enabled']=True
ff_uncropped_p['detection_model'] = 'dlib'
ff_uncropped_p['image_type'] = 'cropped'
ff_uncropped_p=ff_uncropped_p[['file','pred_race', 'pred_gender', 'pred_age_grp',
       'pred_age_lower', 'pred_age_upper', 'pred_model', 'detection_enabled',
       'detection_model', 'image_type']]

ff_cropped_np = pd.read_csv(os.path.join('./cropped_ff_p.csv'))
ff_cropped_np=pd.merge(left=cr_src,how='left',right=ff_cropped_np)
ff_cropped_np['detection_enabled']=False
ff_cropped_np['detection_model'] = None
ff_cropped_np['image_type'] = 'cropped'
ff_cropped_np=ff_cropped_np[['file','pred_race', 'pred_gender', 'pred_age_grp',
       'pred_age_lower', 'pred_age_upper', 'pred_model', 'detection_enabled',
       'detection_model', 'image_type']]

ff_uncropped_np = pd.read_csv(os.path.join('./uncropped_ff_np.csv'))
ff_uncropped_np=pd.merge(left=cr_src,how='left',right=ff_uncropped_np)
ff_uncropped_np['detection_enabled']=False
ff_uncropped_np['detection_model'] = None
ff_uncropped_np['image_type'] = 'uncropped'
ff_uncropped_np=ff_uncropped_np[['file','pred_race', 'pred_gender', 'pred_age_grp',
       'pred_age_lower', 'pred_age_upper', 'pred_model', 'detection_enabled',
       'detection_model', 'image_type']]
```
